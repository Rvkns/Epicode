```python
#Iniziamo filtrando le righe in cui la colonna CONTINENT non risulta vuota
continentfiltered = dataset.dropna(subset=['continent'])
```

```python
continentfiltered
#eseguiamo un check dei record esclusi
```

```python
continent_tot_cases = continentfiltered.groupby('continent')['total_cases'].sum()

print(continent_tot_cases)
```

```python
#Creiamo una funzione che acquisisca in input il dataset e due continenti e restituisce dei valori statistici(casi)
def confronto_casi(dataset, continent1, continent2):
    world_tot_cases = dataset['total_cases'].sum()
    
    #Filtriamo i dataset specifici per iu continenti indicati
    dataset_continent1 = dataset[dataset['continent'] == continent1]
    dataset_continent2 = dataset[dataset['continent'] == continent2]
    
    #Creiamo i descrittori statistici dei continenti(Continente 1)
    min_continent1 = dataset_continent1['total_cases'].min()
    max_continent1 = dataset_continent1['total_cases'].max()
    avg_continent1 = dataset_continent1['total_cases'].mean()
    
    #Creiamo i descrittori statistici dei continenti(Continente 2)
    min_continent2 = dataset_continent2['total_cases'].min()
    max_continent2 = dataset_continent2['total_cases'].max()
    avg_continent2 = dataset_continent2['total_cases'].mean()
    
    #Calcoliamo la percentuale per i continenti indicati
    perc_continent1 = (avg_continent1 / world_tot_cases) 
    perc_continent2 = (avg_continent2 / world_tot_cases) 
    
    results = {
        "Min_{}".format(continent1): min_continent1,
        "Max_{}".format(continent1): max_continent1,
        "Avg_{}".format(continent1): avg_continent1,
        "Perc._{} of World".format(continent1): perc_continent1,
        "Min_{}".format(continent2): min_continent2,
        "Max_{}".format(continent2): max_continent2,
        "Avg_{}".format(continent2): avg_continent2,
        "Perc._{} of World".format(continent2): perc_continent2,
    }
    
    return results


results = confronto_casi(continentfiltered, 'Asia', 'Europe')
print(results)
```

```python
#Creiamo una funzione che acquisisca in input il dataset e due continenti e restituisce dei valori statistici(vaccinazioni)
def confronto_vac(dataset, cont1, cont2):
    world_tot_cases2 = dataset['total_vaccinations'].sum()
    
    dataset_cont1 = dataset[dataset['continent'] == cont1]
    dataset_cont2 = dataset[dataset['continent'] == cont2]
    
    min_cont1 = dataset_cont1['total_vaccinations'].min()
    max_cont1 = dataset_cont1['total_vaccinations'].max()
    avg_cont1 = dataset_cont1['total_vaccinations'].mean()
    
    min_cont2 = dataset_cont2['total_vaccinations'].min()
    max_cont2 = dataset_cont2['total_vaccinations'].max()
    avg_cont2 = dataset_cont2['total_vaccinations'].mean()
    
    perc_cont1 = (avg_cont1 / world_tot_cases2) 
    perc_cont2 = (avg_cont2 / world_tot_cases2) 
    
    results = {
        "Min_{}".format(cont1): min_cont1 , 
        "Max_{}".format(cont1): max_cont1, 
        "Avg_{}".format(cont1): avg_cont1, 
        "Perc._{} of World".format(cont1): perc_cont1,
        "Min_{}".format(cont2): min_cont2,
        "Max_{}".format(cont2): max_cont2,
        "Avg_{}".format(cont2): avg_cont2,
        "Perc._{} of World".format(cont2): perc_cont2,
    }
    
    return results


results = confronto_vac(continentfiltered, 'Asia', 'Europe')
print(results)
```
