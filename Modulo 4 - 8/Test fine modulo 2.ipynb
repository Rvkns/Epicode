```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from plotly import express as px
```

```python
election = px.data.election()
```

```python
election.columns
```

```python
election.rename(columns={'district':'District', 'total':'Total','winner':'Winner','result':'Result','district_id':'District_ID'},
                inplace=True)
```

```python
election.columns
```

```python
district = election['District'][:4]
```

```python
district
```

```python
coderre = election['Coderre'].sum()
bergeron = election['Bergeron'].sum()
joly = election['Joly'].sum()

candidati = ['Coderre', 'Bergeron', 'Joly']
voti = [coderre, bergeron, joly]

plt.figure(figsize=(10,8))
plt.bar(candidati, voti, color=['blue','green','orange'])
plt.xlabel('Candidati')
plt.ylabel('Voti totali')
plt.title('Confronto voti dei candidati')
plt.grid(axis='y')
plt.tight_layout()
plt.savefig('Confronto voti totali', dpi=300)

plt.show()
```

```python
district = election[:4]
cand = ['Coderre', 'Bergeron', 'Joly']
votes = [district['Coderre'].sum(), district['Bergeron'].sum(), district['Joly'].sum()]

plt.figure(figsize=(10,8))
plt.bar(cand, votes, color=['Blue','Green','Orange'])
plt.xlabel('Candidati')
plt.ylabel('Voti totali nei primi 4 distretti')
plt.title('Confronto voti per i primi 4 distretti')
plt.grid(axis='y')
plt.tight_layout()
plt.savefig('Voti per i primi 4 distretti.png', dpi=300)
plt.show()
```

```python
district = election[:4]

# Definiamo i candidati e i voti per i primi quattro distretti
cand = ['Coderre', 'Bergeron', 'Joly']
voti_coderre = district['Coderre'].tolist()
voti_bergeron = district['Bergeron'].tolist()
voti_joly = district['Joly'].tolist()


plt.figure(figsize=(10, 8))
plt.bar(range(len(district)), voti_coderre, color='Blue', label='Coderre')
plt.bar(range(len(district)), voti_bergeron, bottom=voti_coderre, color='Green', label='Bergeron')
plt.bar(range(len(district)), voti_joly, bottom=[voti_coderre[i] + voti_bergeron[i] for i in range(len(district))], color='Orange', label='Joly')

#Abbiamo creato letteralmente tre grafici separati che si sovrappongono grazie al parametro 'Bottom'
#Utilizziamo il contenuto della variabile 'district' come metro di lunghezza del range su cui costruire il grafico, e grazie
#all'iterazione del ciclo FOR serve a calcolare da dove far partire la successiva barra, calcolando la somma di ogni indice 'i'
#presente nel range dei primi quattro distretti

plt.xlabel('Distretti')
plt.ylabel('Voti totali')
plt.title('Confronto voti per i primi 4 distretti')
plt.xticks(range(len(district)), district['District'], rotation=45)
plt.legend()
plt.grid(axis='y')
plt.tight_layout()
plt.savefig('Grafico Impilato.png', dpi=300)

plt.show()
```
