```python
import matplotlib.pyplot as plt
from plotly import express as px
```

```python
stocks = px.data.stocks()
```

```python
goog = stocks['GOOG'][:5]
data = stocks['date'][:5]
#immagaziniamo nelle due variabili di riferimento soltanto le prime cinque righe delle colonne indicate
```

```python
#creiamo un grafico per i primi cinque records, impostando:
plt.figure(figsize=(10, 6)) #parametri di grandezza del grafico
plt.plot(data, goog, marker='*', linestyle='-') #parametri del grafico(dati, stile del marker, stile della linea)
plt.xlabel('Periodo di riferimento') 
plt.ylabel('Prezzo azioni Google')
plt.title('Andamento delle azioni di Google(Primi cinque records)')
#i vari titoli dell'asse X, Y e del grafico generale
plt.xticks(rotation=90) #orientiamo a 90Â° i nomi dell'asse X per evitare sovrapposizioni
plt.grid(False) #eliminiamo la griglia che potrebbe risultare scomoda
plt.show() #mostriamo il grafico
```

```python
goog2 = stocks['GOOG'][-5:]
data2 = stocks['date'][-5:]
#immagaziniamo gli ultimi cinque records in due variabili diverse, per evitare di sovrascrivere quelle precedenti
```

```python
goog2
```

```python
plt.figure(figsize=(10, 6))
plt.plot(data2, goog2, marker='s', linestyle='-')
plt.xlabel('Periodo di riferimento')
plt.ylabel('Prezzo azioni Google')
plt.title('Andamento delle azioni di Google(Ultimi cinque records)')
plt.xticks(rotation=90)
plt.grid(False)
plt.show()
```

```python
azioni = stocks.columns[1:]
azioni
```

```python
plt.figure(figsize=(10,6))
colors = ['red','blue','green','yellow','purple','black','orange']
for col,titoli in enumerate(azioni):
    plt.plot(stocks['date'], stocks[titoli], linestyle='-', linewidth=2, color=colors[col], markerfacecolor="Black")
#facciamo iterare due variabili, una per visualizzare tutti i titoli ed una per cambiargli colore.
#in questo caso abbiamo usato la colonna 'date' come riferimento univoco per i periodi storici e la variabile 'titoli' 
#che contiene l'iterazione delle azioni precedentemente selezionate
plt.xlabel('Periodo di riferimento')
plt.ylabel('Valore delle azioni')
plt.title('Andamento azioni')
plt.xticks(rotation=90, fontsize=5.5) #cambiamo il font dei ticks dell'asse X per evitare 
plt.grid(False)
plt.legend(azioni, loc='lower right', fontsize=8)
plt.tight_layout()

plt.show()
    
```
