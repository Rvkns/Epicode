```python
x = sns.jointplot(data=titanic, x='fare', y='age', kind='scatter')
x.set_axis_labels('Tariffa del biglietto', 'EtÃ  del passeggero')
plt.show()

#In questo modo possiamo osservare la distribuzione dell'etÃ  in relazione al costo del biglietto.
#Si puÃ² notare con grande facilitÃ  che la maggior parte dei passeggeri non erano particolarmente benestanti e che solo una
#piccola parte di questi si Ã¨ potuta permettere un biglietto dal costo superiore ai 50-60 dollari.
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=titanic, x='class')
plt.xlabel('Classe di imbarco')
plt.ylabel('Numero di passeggeri')
plt.title('Passeggeri divisi per classe di imbarco')
plt.grid(axis='y')
plt.show()

#Questo grafico ci permette di visualizzare numericamente la differenza di passeggeri divisi per classe di imbarco.
#E' evidente che i passeggeri in terza classe superano di piÃ¹ del doppio quelli di seconda e prima classe.
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=titanic, x='alive')
plt.xlabel('Sopravvissuto')
plt.ylabel('Numero di passeggeri')
plt.title('Differenza tra il numero di passeggeri sopravvissuti e non sopravvissuti')
plt.grid(axis='y')
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.histplot(data=titanic, x='fare', bins=30, kde=True)
plt.xlabel('Tariffa')
plt.ylabel('Totale')
plt.title('Distribuzione delle tariffe')
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.boxplot(data=titanic, x='pclass', y='age')
plt.xlabel('Classe di imbarco')
plt.ylabel('EtÃ ')
plt.title('Distriuzione delle etÃ  in relazione alla classe di imbarco(Boxplot)')
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.violinplot(data=titanic, x='pclass', y='age')
plt.xlabel('Classe di imbarco')
plt.ylabel('EtÃ ')
plt.title('Distriuzione delle etÃ  in relazione alla classe di imbarco(Violinplot)')
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.swarmplot(data=titanic, x='pclass', y='age')
plt.xlabel('Classe di imbarco')
plt.ylabel('EtÃ ')
plt.title('Distribuzione delle etÃ  in relazione alla classe di imbarco(Swarmplot)')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.kdeplot(data=titanic['age'], shade=True)
plt.xlabel('EtÃ ')
plt.ylabel('DensitÃ ')
plt.title('Distribuzione delle etÃ  (Kernel Density Plot)')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data=titanic, x='survived', y='fare')
plt.xlabel('Sopravvissuto')
plt.ylabel('Tariffa')
plt.title('Distribuzione delle tariffe rispetto alla sopravvivenza')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.lmplot(data=titanic, x='fare', y='survived', logistic=True)
#il parametro logistic ci permette di applicare una regressione lineare logistica.
plt.xlabel('Tariffa')
plt.ylabel('Sopravvissuto')
plt.title('Relazione tra tariffe e sopravvivenza')
plt.show()
```

```python
#Grazie allo studio dei grafici ottenuti possiamo constatare che le persone con un posto privilegiato(Prima e seconda classe)
#hanno avuto maggiori probabilitÃ  di sopravvivenza rispetto a persone che invece hanno viaggiato in terza classe.
```
